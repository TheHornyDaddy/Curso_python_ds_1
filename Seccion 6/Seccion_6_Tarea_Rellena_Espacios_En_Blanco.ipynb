```python
# Define el estilo
sns.set(style="darkgrid", palette="muted", color_codes=True)

# Gráfica los boxplots
ax = sns.boxplot(data=peliculas3, x='Genero', y='Ingresos % USA',\
                  orient='v', color='lightgray', showfliers=False)
plt.setp(ax.artists, alpha=0.5) # aplha=transparency

# Agrega los puntos para mostrar cada observación
sns.stripplot(x='Genero', y='Ingresos % USA', data=peliculas3,\
              jitter=True, size=6, linewidth=0, hue = 'Estudio',\
              alpha=0.7) # jitter agrega etiquetas de manera pseudoaleatoria

ax.axes.set_title('Ingresos Brutos (%) Domésticos por Genero',fontsize=25)
ax.set_xlabel('Genero',fontsize=20)
ax.set_ylabel('Ingreso USA',fontsize=20)

# Define el lugar de la leyenda
ax.legend(bbox_to_anchor=(1.05, 1), loc=2)
plt.show()
```
