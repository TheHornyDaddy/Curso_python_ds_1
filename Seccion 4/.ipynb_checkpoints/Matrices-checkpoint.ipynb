```python
def mi_grafico(lista_jugadores):
    colores = {"KobeBryant":'Black',"JoeJohnson":'Red',"LeBronJames":'Green',"CarmeloAnthony":'Blue',"DwightHoward":'Magenta',"ChrisBosh":'Black',"ChrisPaul":'Red',"KevinDurant":'Green',"DerrickRose":'Blue',"DwayneWade":'Magenta'}
    for nombre in lista_jugadores:
        plt.plot(Games[Pdict[nombre]] , c = colores[nombre] , ls = '--' , marker = '^' , ms = 7 , label = nombre)
    plt.legend(loc = 'upper left' , bbox_to_anchor = (1,1))
    plt.xticks(list(range(0,10)), Seasons , rotation = 'vertical')
    plt.show()
```
