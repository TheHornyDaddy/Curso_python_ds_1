```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams['figure.figsize']=9,4
import warnings
warnings.filterwarnings('ignore')

def mi_grafico(matriz , lista_jugadores = jugadores): # Tomará por default la lista Players
    marcadores = {"KobeBryant":'s',"JoeJohnson":'o',"LeBronJames":'^',"CarmeloAnthony":'D',"DwightHoward":'s',"ChrisBosh":'o',"ChrisPaul":'^',"KevinDurant":'D',"DerrickRose":'s',"DwayneWade":'o'}
    colores = {"KobeBryant":'Black',"JoeJohnson":'Red',"LeBronJames":'Green',"CarmeloAnthony":'Blue',"DwightHoward":'Magenta',"ChrisBosh":'Black',"ChrisPaul":'Red',"KevinDurant":'Green',"DerrickRose":'Blue',"DwayneWade":'Magenta'}
    for nombre in lista_jugadores:
        plt.plot(matriz[Pdict[nombre]] , c = colores[nombre] , ls = '--' , marker = marcadores[nombre] , ms = 7 , label = nombre)
    plt.legend(loc = 'upper left' , bbox_to_anchor = (1,1))
    plt.xticks(list(range(0,10)), temporadas , rotation = 'vertical')
    plt.show()
```

```python
# Tiros Libres Intentados Por Juego
mi_grafico(Tiros_Libres_Intentados / Games)
```

```python
# Precisión de Tiros Libres
mi_grafico(Tiros_Libres / Tiros_Libres_Intentados)
```

```python
# Estilo de Juego del Jugador (preferencia de 2 vs 3 puntos) excluyendo
# los Tiros Libres*
mi_grafico(Points - Tiros_Libres / FieldGoals)
```
